# Imports

```python
import pyod
import matplotlib.pyplot as plt
from pyod.models.ocsvm import OCSVM
from sklearn.metrics import balanced_accuracy_score, roc_auc_score
from pyod.models.deep_svdd import DeepSVDD
import scipy
import sklearn
from sklearn.pipeline import Pipeline
```

# Exercise 1

## 1.

```python
X_train, X_test, y_train, y_test = pyod.utils.data.generate_data(
    n_train=300,
    n_test=200,
    n_features=3,
    contamination=0.15
)
```

```python
fig = plt.figure()
ax = fig.add_subplot(121, projection="3d")
ax.scatter(X_train[:, 0], X_train[:, 1], X_train[:, 2], edgecolors="k", c=y_train, cmap="coolwarm")
ax.set_title("Train data")

ax2 = fig.add_subplot(122, projection="3d")
ax2.scatter(X_test[:, 0], X_test[:, 1], X_test[:, 2], edgecolors="k", c=y_test, cmap="coolwarm")
ax2.set_title("Test data")
plt.show()
```

## 2.

```python
model = OCSVM(
    kernel="linear",
    contamination=0.15
)
model.fit(X_train)

y_train_pred = model.labels_
y_test_pred = model.predict(X_test)

ba_score_test = balanced_accuracy_score(y_test, y_test_pred)
roc_auc_score_test = roc_auc_score(y_test, y_test_pred)

print(f"Balanced accuracy score: {ba_score_test}")
print(f"ROC AUC: {roc_auc_score_test}")
```

## 3

```python
fig = plt.figure()
ax = fig.add_subplot(221, projection="3d")
ax.scatter(X_train[:, 0], X_train[:, 1], X_train[:, 2], edgecolors="k", c=y_train, cmap="coolwarm")
ax.set_title("Train data groud truth")

ax2 = fig.add_subplot(222, projection="3d")
ax2.scatter(X_test[:, 0], X_test[:, 1], X_test[:, 2], edgecolors="k", c=y_test, cmap="coolwarm")
ax2.set_title("Test data ground truth")

ax2 = fig.add_subplot(223, projection="3d")
ax2.scatter(X_train[:, 0], X_train[:, 1], X_train[:, 2], edgecolors="k", c=y_train_pred, cmap="coolwarm")
ax2.set_title("Test data ground predicted")

ax2 = fig.add_subplot(224, projection="3d")
ax2.scatter(X_test[:, 0], X_test[:, 1], X_test[:, 2], edgecolors="k", c=y_test_pred, cmap="coolwarm")
ax2.set_title("Test data ground predicted")

plt.tight_layout()
plt.show()
```

```python
params = [{
    "ocsvm__kernel": ["linear", "poly", "rbf"], 
    "ocsvm__gamma": ["scale", "auto"], 
    "ocsvm__nu": [0.2, 0.5, 0.7
]}]

svm_pipe = Pipeline([
    ('scaler', sklearn.preprocessing.StandardScaler()), 
    ('ocsvm', sklearn.svm.OneClassSVM()),
])

gs_svm = sklearn.model_selection.GridSearchCV(
    estimator=svm_pipe,
    param_grid=params,
    scoring="balanced_accuracy",
    cv=5
)

gs_svm.fit(X_train, y_train)

print(gs_svm.best_params_)
print(f"Best train accuracy: {gs_svm.score(X_train, y_train)}")
```

# Exercise 3

## 1.

```python
data = scipy.io.loadmat("shuttle.mat")
X = data["X"]
y = data["y"].flatten()

# Normalize data 
scaler = sklearn.preprocessing.StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.5)
```

```python
X.shape
```

## 2.

```python
model = OCSVM(
    kernel="rbf",
    contamination=0.2
)
model.fit(X_train)

y_train_pred = model.labels_
y_test_pred = model.predict(X_test)

ba_score_test = balanced_accuracy_score(y_test, y_test_pred)
roc_auc_score_test = roc_auc_score(y_test, y_test_pred)

print(f"Balanced accuracy score: {ba_score_test}")
print(f"ROC AUC: {roc_auc_score_test}")
```

```python
model = DeepSVDD(
    n_features=9,
    contamination=0.2
)
model.fit(X_train)

y_train_pred = model.labels_
y_test_pred = model.predict(X_test)

ba_score_test = balanced_accuracy_score(y_test, y_test_pred)
roc_auc_score_test = roc_auc_score(y_test, y_test_pred)

print(f"Balanced accuracy score: {ba_score_test}")
print(f"ROC AUC: {roc_auc_score_test}")
```

## 3.

```python
model = DeepSVDD(
    n_features=9,
    contamination=0.2,
    hidden_neurons=[64, 32],
    hidden_activation="relu",
    dropout_rate=0.2,
    epochs=100,
    batch_size=64,
)

model.fit(X_train)

y_train_pred = model.labels_
y_test_pred = model.predict(X_test)

ba_score_test = balanced_accuracy_score(y_test, y_test_pred)
roc_auc_score_test = roc_auc_score(y_test, y_test_pred)

print(f"Balanced accuracy score: {ba_score_test}")
print(f"ROC AUC: {roc_auc_score_test}")
```

```python
model = DeepSVDD(
    n_features=9,
    contamination=0.2,
    hidden_neurons=[128, 64, 32, 16],
    hidden_activation="relu",
    dropout_rate=0.3,
    epochs=150,
    batch_size=32,
)

model.fit(X_train)

y_train_pred = model.labels_
y_test_pred = model.predict(X_test)

ba_score_test = balanced_accuracy_score(y_test, y_test_pred)
roc_auc_score_test = roc_auc_score(y_test, y_test_pred)

print(f"Balanced accuracy score: {ba_score_test}")
print(f"ROC AUC: {roc_auc_score_test}")
```
